# Seaborn basics


# univariate anlaysis-when only one feature is present in dataset. Example:f(x)=x1
# Bivariate anlaysis: two features are present. f(x)=x1+x2
# Multivariate analysis: more than two features are present. f(x)=x1+x2+x3

# Distribution plots

# distplot
# joinplot
# pairplot

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.info()
```

# Correlation


```python
# Correlation among features of dataset
df.corr()#correlation will be shown only between features having datatpe as float and int
```

```python
sns.heatmap(df.corr())
```

# Join Plot-helpful in plotting 2d plot-bivariate analysis

```python

sns.jointplot(x='tip', y='total_bill',data=df, kind='hex' )#considering two feature tip & total_bill
```

```python
sns.jointplot(x='tip', y='total_bill',data=df, kind='scatter' )#considering kind as scatter plot
```

# Pair plot-for analyzing more than two features-plot int/float type dtype features

```python
#plotting pairplot
sns.pairplot(df)
```

```python
df['sex'].value_counts()
```

```python
#plotting pairplot classifying any particular feature in scatter plot
sns.pairplot(df,hue='sex')#we use hue parameter to select feature
```

# Dist Plot-to get the distribution of feture

```python
sns.distplot(df['tip'])#showing density on y-axis
```

```python
sns.distplot(df['tip'],kde=False)#after making kde false showing count on y-axis
```

# categorical plot

Seaborn is also helpful in plotting categorical features
boxplot
violinplot
countplot
bar plot

# count plot

```python
#plotting count of different features
sns.countplot(x='sex',data=df)
```

```python
#plotting horizontally
sns.countplot(y='smoker',data=df)
```

# Box plot

```python
#plotting box plot between two features
sns.boxplot(x='smoker',y='tip',data=df)
```

```python
#plotting box plot based on some other feture
sns.boxplot(x='day',y='total_bill',data=df,hue='smoker')
```

# violine plot

```python
sns.violinplot('day','total_bill',data=df,hue='smoker')
```
